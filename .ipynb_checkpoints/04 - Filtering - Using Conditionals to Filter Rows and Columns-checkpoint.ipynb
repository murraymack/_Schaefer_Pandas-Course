```python
# https://www.youtube.com/watch?v=W9XjRYFkkyw&list=PL-osiE80TeTsWmV9i9c58mdDCSskIFdDS&index=3
import pandas as pd
import os

#data_path, schema_path = 'data/news.csv', 'data/news_schema.csv'
#df, schema_df = pd.read_csv(data_path), pd.read_csv(schema_path)
test_dict = {
    "first": ["Dave", "Jane", "John", "Alfred"],
    "last": ["Zimmer", "Doe", "Doe", "Wayne"],
    "email": ["DaveZimmer@mail.com", "JaneDoe@mail.com", "JohnDoe@mail.com", "Alfie@butlers.com"]}
df = pd.DataFrame(test_dict)
```

```python
df
```

```python
df['last'] == 'Doe'
```

```python
filt = (df['last'] == 'Doe')
```

```python
filt
```

```python
df[filt] # df[df['last'] == 'Doe']     <---- This also works
```

```python
df.loc[filt]
```

```python
df.loc[filt, 'email']
```

### AND OR Operators: & |

```python
filt = (df['last'] == 'Doe')  &  (df['first'] == 'John')
df.loc[filt]
```

```python
filt = (df['last'] == 'Doe')  &  (df['first'] != 'John')
df.loc[filt]
```

```python
filt = (df['last'] == 'Zimmer')  |  (df['first'] == 'John')
df.loc[filt]
```

```python
df.loc[~filt]
```
